```python
import anthropic
import os
import json


# Load all secrets first
with open("/Users/sagar/dev/TheBrain/secrets.json") as f:
    secrets = json.load(f)
# Anthropic
client = anthropic.Anthropic(api_key=secrets["ANTHROPIC_API_KEY"])

# Point Claude at your local MCP server
response = client.beta.messages.create(
    model="claude-sonnet-4-20250514",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "What have we been cooking most this month?"}
    ],
    mcp_servers=[
        {
            "type": "stdio",
            "command": "python",
            "args": ["../mcp_server/ourbrain_server.py"],
        }
    ],
    betas=["mcp-client-2025-04-04"]
)

print(response.content[0].text)
```
